```python
import matplotlib.pyplot as plt
import numpy as np

# blue = "#466eff"
# green = "#64B478"
# red = "#EB423D"
# orange = "#FF883D"
# color_list = [blue, green, red, orange]
color_list = [
    "#466eff", "#66CCEE", "#64B478",
    "#88DD44", "#EB423D", "#77423D",
    "#BBBBBB", "#000000", "#FF883D"
]
```

```python
# Ablation for activation function
params = [1e5, 1e6, 1e7]
activation_list=['relu', 'tanh', 'elu', 'selu', 'gelu', 'leakyrelu', 'softpls', 'silu', 'none']
performance = [
    [88.25268, 82.49993, 77.19228],
    [87.01748],
    [86.03512],
    [85.30403],
    [],
    [],
    [],
    [],
    [],
    [],
]

# Create the plot
fig, ax = plt.subplots(figsize=(4, 3))
for idx, perf in enumerate(performance):
	plt.plot(
		params, perf, 'o-', label=activation_list[idx], 
		color=color_list[idx], linewidth=2, markersize=8
	)

# Customize the plot
plt.title("Activation ablation in PPBR_AZ")
plt.xlabel('Number of Parameters', fontsize=12)
plt.ylabel('MAE', fontsize=12)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)

ax = plt.gca()
ax.set_xscale("log")
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x/1e6:.1f}M'))

plt.tight_layout()
plt.show()
```
