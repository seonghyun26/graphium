```python
import matplotlib.pyplot as plt
import numpy as np

blue = "#466eff"
green = "#64B478"
red = "#EB423D"
orange = "#FF883D"
```

```python
# Data for precision 16 (3 points)
params_16 = [1e5, 1e6, 1e7]
performance_16 = [4.6684, 0.6013, 0.50063]  # Performance values
# Data for precision 32 (3 points)
params_32 = [1e5, 1e6, 1e7]
performance_32 = [3.9711, 0.60088, 0.50087]  # Performance values

# Create the plot
fig, ax = plt.subplots(figsize=(4, 3))
plt.plot(
    params_16, performance_16, 'o-', label='Precision 16', 
	color=blue, linewidth=2, markersize=8
)
plt.plot(
    params_32, performance_32, 's-', label='Precision 32', 
	color=green, linewidth=2, markersize=8
)

# Customize the plot
plt.title("Caco2_wang & GCNs")
plt.xlabel('Number of Parameters', fontsize=12)
plt.ylabel('MAE', fontsize=12)
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)

ax = plt.gca()
ax.set_xscale("log")
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, p: f'{x/1e6:.1f}M'))

plt.tight_layout()
plt.show()
```
